```python
# 可选：若容器里缺包，先运行下面一行（装过会自动跳过）
# %pip install -q pandas numpy statsmodels matplotlib

import os, glob
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

OUTDIR = "./outputs"
os.makedirs(OUTDIR, exist_ok=True)

# —— 按你目录里的实际文件名填（与你截图一致）——
PANEL_PATH = "ETH_AwashKesem_Panel_Kebele_NDVI_Rain_TreatPost_2018_2024_v01.csv"
NDVI_PATH  = "ETH_AwashKesem_NDVImean_byKebele_2018_2024_v01.csv"
RAIN_PATH  = "ETH_AwashKesem_RAINJJASmean_byKebele_2018_2024_v01.csv"
VIIRS_PATH = "ETH_AwashKesem_VIIRSmean_byKebele_2018_2024_v01.csv"  # 若没有可注释
COV_PATH   = "ETH_AwashKesem_StaticCovariates_Kebele_SRTM_Slope_v01.csv"  # 可选
```

```python
def load_panel(path, post_year=2020):
    df = pd.read_csv(path)
    # 映射常见列名（容错）
    cmap = {
        "Kebele_ID":   next((c for c in df.columns if c.lower() in ["kebele_id","kebeleid","id","globalid"]), None),
        "Kebele_Name": next((c for c in df.columns if "kebele_name" in c.lower() or c.upper()=="NAME"), None),
        "Year":        next((c for c in df.columns if c.lower()=="year"), None),
        "NDVI":        next((c for c in df.columns if c.upper()=="NDVI"), None),
        "RAIN_JJAS":   next((c for c in df.columns if c.upper()=="RAIN_JJAS"), None),
        "treated":     next((c for c in df.columns if c.lower()=="treated"), None),
        "post":        next((c for c in df.columns if c.lower()=="post"), None),
        "treat_share": next((c for c in df.columns if "treat_share" in c.lower()), None),
        "area_km2":    next((c for c in df.columns if "area_km2" in c.lower()), None),
        "W_NAME":      next((c for c in df.columns if c.upper()=="W_NAME"), None),
        "Z_NAME":      next((c for c in df.columns if c.upper()=="Z_NAME"), None),
        "T_NAME":      next((c for c in df.columns if c.upper()=="T_NAME"), None),
    }
    need = ["Kebele_ID","Year","NDVI","RAIN_JJAS","treated"]
    miss = [k for k in need if cmap[k] is None]
    if miss:
        raise ValueError(f"CSV 缺少关键列：{miss}\n现有列：{list(df.columns)}")

    keep = [v for v in cmap.values() if v is not None]
    df = df[keep].rename(columns={v:k for k,v in cmap.items() if v is not None})

    # 类型与简洁清理
    for c in ["Year","treated","post"]:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce").astype("Int64")
    for c in ["NDVI","RAIN_JJAS","treat_share","area_km2"]:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    if "post" not in df.columns or df["post"].isna().all():
        df["post"] = (df["Year"] >= post_year).astype("Int64")

    # 温和 winsorize（稳健）
    for c in ["NDVI","RAIN_JJAS"]:
        q1, q99 = df[c].quantile([0.01,0.99])
        df[c] = df[c].clip(q1, q99)

    df = df.dropna(subset=["Kebele_ID","Year","NDVI","RAIN_JJAS","treated","post"])
    df["Kebele_ID"] = df["Kebele_ID"].astype(str)
    return df

panel = load_panel(PANEL_PATH, post_year=2020)

print("kebele 数：", panel["Kebele_ID"].nunique())
print("年 度：", sorted(panel["Year"].unique()))
print("总行数：", len(panel))
panel.head()
```

```python
# 处理组/对照组 年度均值（NDVI）
g = (panel.groupby(["treated","Year"])["NDVI"]
          .mean().reset_index()
          .pivot(index="Year", columns="treated", values="NDVI")
          .rename(columns={0:"Control",1:"Treated"}))
display(g.round(4))

# 趋势图（默认配色；单图）
plt.figure()
plt.plot(g.index, g["Control"], marker="o", label="Control")
plt.plot(g.index, g["Treated"], marker="o", label="Treated")
plt.axvline(2020, linestyle="--")
plt.title("Mean NDVI by Year (Treated vs Control)")
plt.xlabel("Year"); plt.ylabel("NDVI"); plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTDIR,"trend_ndvi_treat_vs_control.png"), dpi=200)
plt.show()
```

```python
# ============= 基础包
import os, pandas as pd, numpy as np
import statsmodels.formula.api as smf

# ============= 1) 读取“面板表” + 协变量 + 单位标签（都在同目录）
PANEL_CSV = "ETH_AwashKesem_Panel_Kebele_NDVI_Rain_TreatPost_2018_2024_v01.csv"
COV_CSV   = "ETH_AwashKesem_StaticCovariates_Kebele_SRTM_Slope_v01.csv"
UNIT_CSV  = "ETH_AwashKesem_KebeleUnits_TreatControl_v01.csv"  # 可选（含 treated 与 treat_share）

panel = pd.read_csv(PANEL_CSV)
cov   = pd.read_csv(COV_CSV)
units = pd.read_csv(UNIT_CSV)

# ============= 2) 字段名对齐（保险做法）
# 你关键用到的列：Kebele_ID, Kebele_Name, Year, NDVI, RAIN_JJAS, treated, post, treat_share
need_cols = ["Kebele_ID","Kebele_Name","Year","NDVI","RAIN_JJAS","treated","post","treat_share"]
print("panel columns:", list(panel.columns))

# 有些导出的行政层级列名可能是 W_NAME/Z_NAME/T_NAME；并不影响回归
# 统一 ID/名称的类型
for c in ["Kebele_ID","Kebele_Name"]:
    if c in panel.columns:
        panel[c] = panel[c].astype(str)

# ============= 3) 数值化与类型清洗（关键：去掉 pandas 的可空整型 Int64Dtype）
to_float = ["NDVI","RAIN_JJAS"]
to_int   = ["Year","treated","post"]

for c in to_float:
    if c in panel.columns:
        panel[c] = pd.to_numeric(panel[c], errors="coerce").astype(float)

for c in to_int:
    if c in panel.columns:
        panel[c] = pd.to_numeric(panel[c], errors="coerce")
        # 缺失年份/处理标记先丢掉；也可以按需要填补
        panel = panel[~panel[c].isna()]
        panel[c] = panel[c].astype("int64")   # ← 变成 numpy int64，patsy 可识别

# treat_share 可能是浮点
if "treat_share" in panel.columns:
    panel["treat_share"] = pd.to_numeric(panel["treat_share"], errors="coerce").astype(float)

# 合并静态协变量（可选）
if set(["elevation","slope"]).issubset(cov.columns):
    cov_sel = cov[["Kebele_ID","elevation","slope"]].copy()
    cov_sel["Kebele_ID"] = cov_sel["Kebele_ID"].astype(str)
    panel = panel.merge(cov_sel, on="Kebele_ID", how="left")

# 如果想用单位表里更“干净”的 treated / treat_share，可覆盖：
for c in ["treated","treat_share"]:
    if c in units.columns:
        panel = panel.drop(columns=[c], errors="ignore").merge(
            units[["Kebele_ID", c]].assign(Kebele_ID=lambda d: d["Kebele_ID"].astype(str)),
            on="Kebele_ID", how="left"
        )
        if c=="treated":
            panel["treated"] = pd.to_numeric(panel["treated"], errors="coerce").fillna(0).astype("int64")
        if c=="treat_share":
            panel["treat_share"] = pd.to_numeric(panel["treat_share"], errors="coerce").astype(float)

# ============= 4) 去异常/缺失
panel = panel.dropna(subset=["NDVI","RAIN_JJAS","treated","post","Year","Kebele_ID"]).copy()

# 设定分类，以便 C() 固定效应更稳
panel["Kebele_ID"] = panel["Kebele_ID"].astype("category")
panel["Year_cat"]  = panel["Year"].astype("category")  # 保留原 Year 供聚合/绘图

print(panel.dtypes)

# ============= 5) DID 回归（statsmodels + 集群稳健标准误）
# 规格：NDVI_it = β (treated_i*post_t) + γ RAIN_it + Year FE + Unit FE
form = "NDVI ~ treated:post + RAIN_JJAS + C(Year_cat) + C(Kebele_ID)"

did = smf.ols(formula=form, data=panel).fit(
    cov_type="cluster",
    cov_kwds={"groups": panel["Kebele_ID"]}  # kebele 聚类稳健
)

print(did.summary().as_text())

# 把核心系数与置信区间输出一个小表
coef = did.params.get("treated:post", np.nan)
ci_l, ci_u = did.conf_int().loc["treated:post"].tolist() if "treated:post" in did.params.index else [np.nan, np.nan]
res_tab = pd.DataFrame({"coef":[coef], "ci_low":[ci_l], "ci_high":[ci_u], "N":[len(panel)]})
display(res_tab)

# ============= 6) （可选）用 linearmodels 的实体/时间固定效应做法（等价）
# 如果你镜像里已装了 linearmodels，解开注释使用（速度通常更好）
# from linearmodels.panel import PanelOLS
# p2 = panel.set_index(["Kebele_ID","Year"]).sort_index()
# exog = pd.DataFrame({"treated_post": (p2["treated"]*p2["post"]).astype(float),
#                      "RAIN_JJAS": p2["RAIN_JJAS"].astype(float)}, index=p2.index)
# mod  = PanelOLS(p2["NDVI"], exog, entity_effects=True, time_effects=True)
# fe   = mod.fit(cov_type="clustered", clusters=p2.index.get_level_values(0))
# print(fe.summary)
```

```python
# ============= 7) Event Study (平行趋势检验)
POST_YEAR = 2020
panel["rel_year"] = panel["Year"].astype(int) - POST_YEAR

# 保留窗口（例如前 3 年到后 4 年，可按数据期改）
LEADS, LAGS = 3, 4
df_es = panel.query("rel_year>=-@LEADS and rel_year<=@LAGS").copy()

# 创建交互项 treated × year_dummies
for k in range(-LEADS, LAGS+1):
    df_es[f"evt_{k}"] = ((df_es["rel_year"]==k) & (df_es["treated"]==1)).astype(int)

# -1 年作为基准
evt_terms = " + ".join([f"evt_{k}" for k in range(-LEADS, LAGS+1) if k!=-1])

form_evt = f"NDVI ~ {evt_terms} + RAIN_JJAS + C(Year_cat) + C(Kebele_ID)"
evt = smf.ols(formula=form_evt, data=df_es).fit(
    cov_type="cluster",
    cov_kwds={"groups": df_es["Kebele_ID"]}
)

# 抽取系数
rows = []
for k in range(-LEADS, LAGS+1):
    if k==-1:
        rows.append({"rel_year":k,"coef":0,"ci_low":0,"ci_high":0})
    else:
        nm = f"evt_{k}"
        co = evt.params.get(nm, np.nan)
        lo,hi = evt.conf_int().loc[nm].tolist() if nm in evt.params.index else (np.nan,np.nan)
        rows.append({"rel_year":k,"coef":co,"ci_low":lo,"ci_high":hi})
es = pd.DataFrame(rows)

# 绘图
plt.figure(figsize=(7,4))
plt.axhline(0, color="k", linewidth=1)
plt.axvline(-0.5, linestyle="--")
plt.errorbar(es["rel_year"], es["coef"], 
             yerr=[es["coef"]-es["ci_low"], es["ci_high"]-es["coef"]],
             fmt="o", capsize=3)
plt.title("Event Study: Treated × Year relative to 2020")
plt.xlabel("Years relative to 2020"); plt.ylabel("Effect on NDVI")
plt.grid(True); plt.tight_layout()
plt.savefig(os.path.join(OUTDIR,"event_study.png"), dpi=200)
plt.show()
```
